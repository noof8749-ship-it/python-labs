# 5.6 del Statement

### Deleting the Element at a Specific List Index

```python
numbers = list(range(0, 10))
```

```python
numbers
```

```python
del numbers[-1]
```

```python
numbers
```

### Deleting a Slice from a List

```python
del numbers[0:2]
```

```python
numbers
```

```python
del numbers[::2]
```

```python
numbers
```

### Deleting a Slice Representing the Entire List

```python
del numbers[:]
```

```python
numbers
```

### Deleting a Variable from the Current Session

```python
del numbers
```

```python
numbers
```
